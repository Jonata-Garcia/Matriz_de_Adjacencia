```python
# Definindo os vrtices
vertices = ["A", "B", "C", "D", "E"]

# Criando a matriz 5x5 inicial com zeros
n = len(vertices)
matriz = [[0 for _ in range(n)] for _ in range(n)]

# Lista de arestas do grafo (no direcionado)
arestas = [
    ("A", "C"),
    ("B", "C"),
    ("C", "D"),
    ("C", "E"),
    ("D", "E")
]

# Preenchendo a matriz de adjacncia
for u, v in arestas:
    i = vertices.index(u)
    j = vertices.index(v)
    matriz[i][j] = 1
    matriz[j][i] = 1   # porque o grafo  no-direcionado

# Exibindo a matriz
print("Matriz de Adjacncia:")
for linha in matriz:
    print(linha)
```
